- - can use hex values for color as well
- to mention linewidth use attribute in plt.plot(linewidth = 3)
- plt.tight_layout() --> for padding issues

### Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]
# plot mein ham array pass karte hain.
plt.plot(x, y)
# title graph ka
plt.title('Our First Graph !', fontdict={
          'fontname': 'Comic Sans MS', 'fontsize': 15})
# x axis aur y axis ka label
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Resize the graph (dpi is pixels per inch. when saving probably should use 300 if possible)
plt.figure(figsize=(5, 3), dpi=100)

# Keyword Argument Notation
# plt.plot(x, y, label='data1',color='red',linewidth=4,linestyle='--',marker='.',markersize=20,markeredgecolor='blue')
# instead of this we can do like

# Use short hand notation
# fmt='[color][marker][line]'
# can check these short hand values on net
plt.plot(x, y, 'b.--', label='2x')

# line 2

# select intervals we want to plot points at
x2 = np.arange(0, 4.5, 0.5)

# plot part of graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')

# plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')

# add a title (specify font parameters with fontdict)
plt.title('Our First Graph !', fontdict={
          'fontname': 'Comic Sans MS', 'fontsize': 15})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Ticks is basically points on graph you want
# X,Y axis tickmarks(scale on your graph)
plt.xticks([0, 1, 2, 3, 4])
# plt.yticks([0, 2, 4, 6, 8])

# add a legend
plt.legend()

# save figure (dpi 300 is good when saving so graph has higher resolution)
plt.savefig('mygraph.png',dpi=300)

# for padding
plt.tight_layout()

# for grid
plt.grid(True)

# show plot
plt.show()
```

### Bar Charts

```python
labels=['A','B','C']
values=[1,4,2]

bars=plt.bar(labels,values)

patterns=['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))


# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

plt.figure(figsize=(3,2))

plt.show()
```
